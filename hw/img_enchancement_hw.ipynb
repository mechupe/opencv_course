```python
import cv2
import matplotlib.pyplot as plt
```

```python
ya_logo = cv2.imread('../data/yandex.jpg')
ya_logo = cv2.cvtColor(ya_logo, cv2.COLOR_BGR2RGB)
ya_logo_gray = cv2.cvtColor(ya_logo, cv2.COLOR_BGR2GRAY)


y_retval, y_img_thresh = cv2.threshold(ya_logo_gray, 10, 255, cv2.THRESH_BINARY)
retval, img_thresh = cv2.threshold(ya_logo, 1, 255, cv2.THRESH_BINARY)

plt.imshow(ya_logo);
plt.show()
plt.imshow(cv2.bitwise_not(img_thresh), cmap='gray');
plt.show()
plt.imshow(y_img_thresh, cmap='gray');
plt.show()
```

```python
mask = cv2.bitwise_not(y_img_thresh)
andex = cv2.bitwise_and(img_thresh, img_thresh, mask=mask)
red_letter = cv2.add(andex, ya_logo)
plt.imshow(red_letter);
```

```python
flag = cv2.imread('../data/flag.jpeg')
flag = cv2.cvtColor(flag, cv2.COLOR_BGR2RGB)
plt.imshow(flag)
```

```python
ratio = red_letter.shape[0] / flag.shape[1]

flag_resized = cv2.resize(flag, (red_letter.shape[1], red_letter.shape[0]))
plt.imshow(flag_resized)
```

```python
masked_flag = cv2.bitwise_and(flag_resized, flag_resized, mask=y_img_thresh)
plt.imshow(masked_flag)
```

```python
final = cv2.add(cv2.bitwise_not(img_thresh), masked_flag)
plt.imshow(final)
```

```python
final_black = cv2.add(masked_flag, andex)
plt.imshow(final_black)
```
